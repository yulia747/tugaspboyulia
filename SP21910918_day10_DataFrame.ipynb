## A. Pengenalan
--------------
**Data Frame** yaitu sebuah koleksi kolom dan baris berurutan dengan memiliki nama dan jenis tertentu, sehingga seperti layaknya sebuah tabel yang tampak seperti database dimana sebuah baris tunggal mewakili sebuah contoh tunggal dan kolom mewakili atribut tertentu. Harus dicatat di sini bahwa elemen dalam berbagai kolom mungkin berapa jenis yang berbeda.
Seperti tipe data Hash dan List, Data Frame bisa menyimpan dengan tipe data berbeda

Berikut ini kegunaan utama dari sebuah tipe frame data. 
* Importing and exporting data from text files and databases
* Adding and removing columns
* Sorting
* Filtering
* Creating new columns by applying functions to existing ones (mapping) 
* Summarizing columns or tables (reducing)
* Combining tables by appending or joining
* Calculating descriptive statistics
* Adding, updating, and removing rows

### C.2 Casting Kolom Table

Casting adalah perubahan type data ke type data lainnya dengan memperhatikan ukuran kapasitas di setiap type data. Type data yang sudah dijelaskan sebelumnya di Bahasa Java dalam Pemrograman.

contohnya adalah dari integer ke float, integer ke long, float ke double, dan seterusnya, maka dalam Casting yang terjadi adalah dari integer ke String, String ke integer, float ke String, String ke float dan masih banyak lagi sesuai dengan ketentuan yang ada di java.
Berikut contoh coding untuk merubah integer ke long:


```javascript

  int nilai= 60;
  String hasilCast = String.valueOf(nilai);//Casting dari int ke String
  System.out.println(hasilCast);

  int hasilCast2 = Integer.valueOf(nilai); // Casting dati String ke int
  System.out.println(hasilCast2);

```

Seperti halnya variable dasar kadang tipe data kolom yang ada tidak sesui dengan tipe data yang kita butuhkan. Sebagai contoh kita ingin melakukan komputasi atau penghitungan dari sebuah kolom yang bertipe *String* atau *char*. Padahal tipe data tersebut kita tidak bisa melakukan perhitungan aritmetik terhadap anggota kolom tersebut. Hal yang dapat kita lakukan adalah melakukan prosedur casting dari kolom tersebut. 



### C.3 Manipulasi Baris Tabel

Sebagaimana **column** dalam data frame disediakan fungsi - fungsi atau method yang digunakan unutuk manipulasi **row** atau baris.

```java

// drop rows using Selections
result = table.dropWhere(table.numberColumn(0).isLessThan(100));

// add rows
destinationTable.addRow(43, sourceTable); // adds row 43 from sourceTable to the receiver

// sampling
table.sampleN(200); // select 200 rows at random from table
```

Kita juga bisa melakukan *arbitrary operations* pada tiap -tiap baris pada sebuah table. Dengan sebuah iterasi kita bisa melakukan operasi terhadap baris seperti contoh berikut.

```java

for (Row row :  Binatang) {
  System.out.println("Baris " + row.getString("AnimalName") + ": " + row.getDouble("Rating"));
}

```

Atau kita bisa mengakses isi data pada tiap cell dari sebuah baris dengan menggunakan iterasi.

```java
table.stream()
    .forEach(
        row -> {
          System.out.println("On " + row.getDate("date") + ": " + row.getDouble("approval"));
        });
```

Berikut methoda yang bisa dilakukan untuk memproses suatu baris atau lebih dalam suatu waktu.
If you need to process more than one row at a time, there are several methods to help. 

```javascript
// Consumer prints out the max of a window.
Consumer<Row[]> consumer =
    rows ->
        System.out.println(Arrays.stream(rows).mapToDouble(row -> row.getDouble(0)).max());

// Streams over rolling sets of rows. I.e. 0 to n-1, 1 to n, 2 to n+1, etc.
table.rollingStream(3).forEach(consumer);

// Streams over stepped sets of rows. I.e. 0 to n-1, n to 2n-1, 2n to 3n-1, etc. Only returns
// full sets of rows.
table.steppingStream(5).forEach(consumer);
```

### C.4 Sorting
Untuk pengurutan (sort) dari suatu Table kita bisa memanggil methoda `sortOn()` dengan parameter colomn.  

```javascript
Table sorted = table.sortOn("foo", "bar", "bam"); // Sorts Ascending by Default
sorted = table.sortAscendingOn("bar"); // just like sortOn(), but makes the order explicit.
sorted = table.sortDescendingOn("foo");

// sort on foo ascending, then bar descending. Note the minus sign preceding the name of
// column bar.
sorted = table.sortOn("foo", "-bar");
```


### C.5 Filtering

Filtering digunakan ketika kita ingin mencari suatu data tertentu dalam sebuah Table. Kita juga bisa mengkombinasikan Query filter methoda dengan operasi logika
seperti *and*, *or*, and *not*. Semua kemapuan tersebut diatas diimplementasikan dalam class `QuerySupport`. 

```javascript
import static tech.tablesaw.api.QuerySupport.and;
import static tech.tablesaw.api.QuerySupport.or;
import static tech.tablesaw.api.QuerySupport.not;
```

Setiap method atau fungsi dalam mengimplentasikan kemampuan tata struktur *Lambadas* dengan sintaks `Function<Table, Selection>` sebagaimana contoh berikut.

```javascript
Table result =
    table.where(
        and(
            or(
                t -> t.doubleColumn("nc1").isGreaterThan(4),
                t -> t.doubleColumn("nc1").isNegative()
                ),
            not(t -> t.doubleColumn("nc2").isLessThanOrEqualTo(5))));
```
